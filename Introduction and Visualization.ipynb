### The QQ Plot

#### To find the density of the plot
- The qq plot is used to find if data is distributed in a certain way. Usually its used to see how the data fits in normal distribution

```python
import scipy.stats as ss
import pylab
```

```python
ss.probplot(data['spx'],plot=pylab)
pylab.show()
```

- y axis contains prices arranged in ascending order
- x axis is theoretical quantiles, how many std away from mean these values are
- red diagonal shows what values should be to be normal distribution
- Therefore, the data is not normally distributed
